```python
import re
def fog_index(text):
    
    # Split text into sentences using regex
    sentences = re.findall(r"([A-Za-z][^.!?]*[.!?])", text)

    # Initialize variables
    num_sentences = len(sentences)
    num_words = 0
    num_complex_words = 0

    # Calculate number of words and complex words
    for sentence in sentences:
        words = sentence.split()
        num_words += len(words)
        num_complex_words += sum([1 for word in words if len(re.findall(r'[aeiou]{2,}', word, re.IGNORECASE)) >= 1])
# Average Number of Words Per Sentence = the total number of words/the total number of sentences
# Average_Number_of_Words_Per_Sentence=num_words/num_sentences

    # Average Sentence Length = the number of words / the number of sentences
    Average_Sentence_Length=num_words/num_sentences

    # Percentage of Complex words = the number of complex words / the number of words 
    Percentage_of_Complex_words=num_complex_words/num_words
    # Calculate Fog index
    fog_index = 0.4 * ((num_words / num_sentences) + (100 * (num_complex_words / num_words)))
    
    Values.append(Average_Sentence_Length)
    Values.append(Percentage_of_Complex_words)
    Values.append(fog_index)
#   Values.append(Average_Number_of_Words_Per_Sentence)
fog_index(info_with)
```

```python
# # Define a function to count the number of syllables in a word
def count_syllables(w):
    vowels = "aeiouy"
    count_ = 0
    for i in range(len(w)):
        if i == 0 and w[i] in vowels:
            count_ += 1
        elif w[i] in vowels and w[i-1] not in vowels:
            count_ += 1
    if w.endswith('e'):
        count_ -= 1
    if count_ == 0:
        count_ += 1
    return count_
# Loop through the words and count their syllables
Complex_Word_Count = []
for w in word_token:
    if count_syllables(w) > 2:
        Complex_Word_Count.append(w)
count_syllables(info_with)
```
